## print() Function
The print function is simple. To exclude the newline after the output, use the 'end' argument:

```python
print("hello")
print("world")
```

```python
print("hello", end=' ')
print("world")
```

The 'sep' keyworld allows the separation of objects if there are more than one:

```python
print("cats","dogs","mice", sep=",")
```

# len() Function
Evaluates to the integer value of the number of characters in a string, list, dictionary, etc.

When testing if a variable is empty, use direct boolean evalution rather than 'len'

```python
len('hello')
```

```python
len(['cat', 3, 'dog'])
```

```python
a = [1, 2, 3]

if a:
    print("not empty")
```
